```python
#W5A1
def Max_of_Two(a, b):
    if a >= b:
        return a
    return b
a, b = map(int, input().split())
Max_of_Two(a, b)
```

```python
#W5A2
def Swap(a, b):
    return (b, a)
a, b = map(int, input().split())
Swap(a, b)
```

```python
#W5A3
import math as m
def IsPrime(n):
    if n < 2:
        return False
    for i in range(2, m.isqrt(n) + 1):
        if n % i == 0:
            return False
    return True
n = int(input())
IsPrime(n)
```

```python
#W5A4
import math as m
def perfect_num(n):
    S = 0
    for i in range(1, m.isqrt(n) + 1):
        if n % i == 0:
            if n // i != i:
                S += i
                S += n // i
            else:
                S += i
    return S == n * 2
n = int(input())
perfect_num(n)
```

```python
#W5A5
def find_k(k, x):
    for i in range(len(x)):
        if x[i] == k:
            return i
    return -1
x = list(map(int, input().split()))
k = int(input())
find_k(k, x)
```

```python
#W5A6
def factorial(n):
    f = 1
    for i in range(1, n + 1):
        f *= i
    return f
n = int(input())
factorial(n)
```

```python
#W5A7
def calc(x, y, op):
    if op == '-':
        return x - y
    if op == '+':
        return x + y
    if op == '/':
        return round(x / y, 2)
    if op == '*':
        return x * y
x, y = map(float, input().split())
op = input()
calc(x, y, op)
```

```python
#W5A8
def Hamming(x, y):
    n = x ^ y
    return n.bit_count()
x, y = map(int, input().split())
Hamming(x, y)
```

```python
#W5A9
def sum_digit(n):
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    return s
n = int(input())
sum_digit(n)
```

```python
#W5A10
def check(x, y):
    mp_ab = [-1] * 26
    mp_ba = [-1] * 26
    for i in range(len(x)):
        Id = ord(x[i]) - 97
        val = ord(y[i]) - 97
        if mp_ab[Id] != -1:
            if mp_ab[Id] != val:
                return False
        else:
            mp_ab[Id] = val
    for i in range(len(x)):
        Id = ord(y[i]) - 97
        val = ord(x[i]) - 97
        if mp_ba[Id] != -1:
            if mp_ba[Id] != val:
                return False
        else:
            mp_ba[Id] = val
    return True
x, y = input().split()
check(x, y)
```
