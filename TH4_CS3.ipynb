```python
#W4A6 - Sum Primes in range
import math
a, b = map(int, input().split())
Sum = 0
for i in range(a, b + 1):
    if i == 1:
        continue
    flag = True
    for j in range(2, math.isqrt(i) + 1):
        if i % j == 0:
            flag = False
            break
    if flag:
        Sum += i
print(Sum)
```

```python
#W4A8 - ReverseAddPalindrome
def reverse(n):
    rev_n = 0
    while n > 0:
        rev_n = rev_n * 10 + n % 10
        n //= 10
    return rev_n
n = int(input())
cnt = 0
while n != reverse(n):
    n += reverse(n)
    cnt += 1
print(f'''so buoc la {cnt}
gia tri thu duoc la {n}''')
```

```python
#W4A9 - PerfectSparesDistinctDigits
import math as m
def check(n):
    dup = [False] * 10
    while n > 0:
        if dup[n % 10] == True:
            return False
        dup[n % 10] = True
        n //= 10
    return True
n = int(input())
i = 1
while (i * i <= n):
    if (check(i * i)):
        print(i * i, end = ' ')
    i += 1
```

```python
#W4A10 - LongestCollatzUptoN
n = int(input())
x = 1
L = 1
for i in range(1, n + 1):
    num = i
    Len = 1
    while num != 1:
        if num % 2 == 0:
            num /= 2
        else:
            num = 3 * num + 1
        Len += 1
    if Len > L:
        x = i
        L = Len
print(x, L)
```

```python
#W4A13 - ClosePairNumber
from math import isqrt
def SumDivisor(n):
    Sum = 0
    for i in range(1, isqrt(n) + 1):
        if n % i == 0:
            Sum += i
            if n // i != i:
                Sum += n/i
    return Sum - n
a, b = map(int, input().split())
if SumDivisor(a) == b and SumDivisor(b) == a:
    print ("true")
else:
    print("false")
```

```python
#W4A14 - GCD
from math import isqrt
try:
    m, n = map(int, input("Nhập 2 số nguyên dương m và n: ").split())
    if (m < 0 or n < 0):
        print("Nhập số nguyên dương!")
    else:
        gcd = 1
        if m == 0 or n == 0:
            gcd = max(m, n)
        else: 
            for i in range(1, isqrt(m) + 1):
                if m % i == 0:
                    if n % i == 0:
                       gcd = max(gcd, i)
                    if n % (m // i) == 0:
                        gcd = max(gcd, m//i)
        print(f'Ước chung lớn nhất của {m} và {n} là {gcd}')
except ValueError:
    print('Vui lòng nhập số nguyên dương!')
```

```python
#W4A15 - ChickenAndDog
m = int(input("Nhập tổng số gà và chó: "))
n = int(input("Nhập tổng số chân gà và chó: "))
if (n - 2 * m) % 2 != 0:
    print('invalid')
else:
    dog = (n - 2 * m) // 2
    chicken = m - dog
    if chicken < 0 or dog < 0:
        print('invalid')
    else:
        print(f'Số gà là {chicken}', f'Số chó là {dog}', sep = '\n')
```
