```python
#Bai4
h = int(input())
for i in range(1, h + 1):
    print (' ' * (h - i), end = '')
    if i == 1:
        print('*')
        continue
    if i == h:
        print('*' * (2 * h - 1))
        continue
    print('*', end = '')
    print(' ' * ((i - 1) * 2 - 1), end = '')
    print('*')
```

```python
#Bai7
import math as m
a = int(input())
cnt = 0
for i in range(1, m.isqrt(a) + 1):
    if (a % i == 0):
        if (a / i != i):
            cnt += 2
        else:
            cnt += 1
print(cnt) 
```

```python
#Bai14
Min = 1e12
Max = -1e12
while True:
    a = int(input())
    if (a == -1):
        break
    Min = min(Min, a)
    Max = max(Max, a)
print(f'Số lớn nhât là {Max}')
print(f'Số bé nhất là {Min}')
```

```python
#Bai15
def CountDigit1(num):
    cnt = 0
    while (num > 0):
        cnt += 1
        num //= 10
    return cnt
def CountDigit2(num):
    n = str(num)
    return len(n)
n = int(input("Nhap so nguyen duong n: "))
print(f'so chu so cua n la {CountDigit1(n)}')
print(f'so chu so cua n la {CountDigit2(n)}')
```

```python
#Bai16
n = int(input("Nhap so nguyen duong n: "))
num = n
even = 0
odd = 0
while num > 0:
    if (num % 10) % 2:
        odd += 1
    else:
        even += 1
    num //= 10
print(f'Số chữ số chẵn của {n} là: {even}')
print(f'Số chữ số lẻ của {n} là: {odd}')
```

```python
#Bai17
def SumDigit(n):
    s = 0;
    while n > 0:
        s += n % 10
        n //= 10
    return s
n = int(input("Nhap so nguyen duong n: "))
print(f'Tổng các chữ số của {n} là: {SumDigit(n)}')
```
