```python
#Bai4
h = int(input())
for i in range(1, h + 1):
    print (' ' * (h - i), end = '')
    if i == 1:
        print('*')
        continue
    if i == h:
        print('*' * (2 * h - 1))
        continue
    print('*', end = '')
    print(' ' * ((i - 1) * 2 - 1), end = '')
    print('*')
```

```python
#Bai7
import math as m
a = int(input())
cnt = 0
for i in range(1, m.isqrt(a) + 1):
    if (a % i == 0):
        if (a / i != i):
            cnt += 2
        else:
            cnt += 1
print(cnt) 
```

```python
#Bai8
a, b = map(int, input().split())
for i in range(1, min(a, b) + 1):
    if (a % i == 0 and b % i == 0):
        print(i, end = ' ')
```

```python
#Bai10
import math as m
a = int(input())
sum = 0;
for i in range(1, m.isqrt(a) + 1):
    if a % i == 0:
        sum += i
        if (a / i != i):
            sum += a / i
if (sum / 2 == a): 
    print(f'{a} la so hoan hao')
else:
    print(f'{a} khong phai la so hoan hao')
    
```
