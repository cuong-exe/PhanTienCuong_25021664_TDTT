```python
#W6A19
def search(a):
    Max = a[0]
    idx = 0
    for i in range(1, len(a)):
        if a[i] > Max:
            Max = a[i]
            idx = i
    return idx

a = list(map(int, input().split()))
print(search(a))
```

```python
#W6A20
def search(a, k):
    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1

a = list(map(int, input().split()))
k = int(input())
print(search(a, k))
```

```python
#W6A22
def vi_tri_max(lst):
    if not lst:
        return -1
    Max = lst[0]
    idx = 0
    for i in range(1, len(lst)):
        if lst[i] > Max:
            Max = lst[i]
            idx = i
    return idx

ds = list(map(int, input().split()))
print(vi_tri_max(ds)) 
```

```python
#W6A23
def xoa_nho_hon_x(lst, x):
    return [num for num in lst if num >= x]

ds = list(map(int, input().split()))
x = int(input())
ans = xoa_nho_hon_x(ds, x)
print(ans)
```

```python
#W6A24
def day_0_ve_dau(lst):
    zeros = [num for num in lst if num == 0]
    others = [num for num in lst if num != 0]
    return zeros + others

def day_0_ve_cuoi(lst):
    zeros = [num for num in lst if num == 0]
    others = [num for num in lst if num != 0]
    return others + zeros

ds = list(map(int, input().split()))

print(day_0_ve_dau(ds))  
print(day_0_ve_cuoi(ds))  
```

```python
#W6A25
def gia_tri_thieu(lst):
    full_set = set(range(10))
    present_set = set(lst)
    missing = full_set - present_set
    return sorted(list(missing))

ds = list(map(int, input().split()))
print(gia_tri_thieu(ds))  # Kết quả: [1, 5, 6, 7, 8]
```

```python
#W6A26
def tong_cuc_dai_dia_phuong(lst):
    if len(lst) < 3:
        return 0
    tong = 0
    for i in range(1, len(lst) - 1):
        if lst[i] > lst[i - 1] and lst[i] > lst[i + 1]:
            tong += lst[i]
    return tong

ds = list(map(int, input().split()))
print(tong_cuc_dai_dia_phuong(ds)) 
```

```python
#W6A27
def vi_tri_nho_hon_x(lst, x):
    positions = []
    for i in range(len(lst)):
        if lst[i] < x:
            positions.append(i)
    if positions:
        return positions
    else:
        return -1

ds = list(map(int, input().split()))
x = 4
print(vi_tri_nho_hon_x(ds, x))
```

```python
#W6A28
def kiem_tra_day(lst):
    if len(lst) < 2:
        return "Không xác định" 
    
    tang = True
    giam = True
    
    for i in range(len(lst) - 1):
        if lst[i] >= lst[i+1]:
            tang = False
        if lst[i] <= lst[i+1]:
            giam = False
    
    if tang:
        return "Dãy tăng thực sự"
    elif giam:
        return "Dãy giảm thực sự"
    else:
        return "Vô danh"
ds = list(map(int, input().split()))
print(kiem_tra_day(ds))     
```
