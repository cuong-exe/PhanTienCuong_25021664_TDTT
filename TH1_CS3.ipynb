```python
#W1A1
# Gán giá trị số nguyên 7 cho biến có tên là 'a'
a = 7 
# Gán giá trị số nguyên 5 cho biến có tên là 'b'
b = 5 
# Thực hiện phép trừ giá trị của 'a' cho 'b' (7 - 5)
# và gán kết quả (là 2) cho biến 'c'
c = a - b 
# In giá trị hiện tại của biến 'c' ra màn hình
print(c)
```

```python
#W1A2
# Khai báo biến 'city' và gán cho nó giá trị chuỗi là "Hà Nội"
city = "Hà Nội" 
# Khai báo biến 'year' và gán cho nó giá trị số nguyên là 2025
year = 2025 
# In ra màn hình một chuỗi được ghép từ các chuỗi và giá trị của các biến.
# Các phần tử trong hàm print được ngăn cách bởi dấu phẩy sẽ được tự động thêm một dấu cách khi in ra.
print("Thành phố: ", city, " - Năm: ", year)
```

```python
#W1A3
# Gán giá trị 4 cho biến n. Biến này đại diện cho giới hạn trên của dãy số.
n = 4
# Khởi tạo biến t (tổng) với giá trị ban đầu là 0.
# Biến này sẽ được dùng để tích lũy tổng của các số.
t = 0
# Bắt đầu vòng lặp for.
# Vòng lặp sẽ duyệt qua các số nguyên từ 1 đến n (bao gồm cả n).
# Hàm range(1, n+1) sẽ tạo ra một chuỗi số 1, 2, 3, 4.
for i in range(1, n+1):
  # Cứ mỗi lần lặp, cộng giá trị hiện tại của i vào biến t.
  # Ví dụ:
  # Lần 1: i = 1, t = 0 + 1 = 1
  # Lần 2: i = 2, t = 1 + 2 = 3
  # Lần 3: i = 3, t = 3 + 3 = 6
  # Lần 4: i = 4, t = 6 + 4 = 10
  t += i
# Sau khi vòng lặp kết thúc, in giá trị cuối cùng của t ra màn hình.
# Kết quả sẽ là 10.
print(t)
```

```python
#W1A4
# Khởi tạo một danh sách (list) các số nguyên
numbers = [1, 2, 3, 4]
# Bắt đầu vòng lặp. Vòng lặp sẽ duyệt qua từng phần tử trong danh sách 'numbers'.
# Mỗi phần tử sẽ được gán cho biến 'x' trong mỗi lần lặp.
for x in numbers:
  # Kiểm tra xem số 'x' có phải là số chẵn không.
  # Toán tử % (chia lấy dư). Nếu x % 2 == 0, nghĩa là x chia hết cho 2.
  if x % 2 == 0:
    # Nếu là số chẵn, in ra thông báo tương ứng.
    print(x, "là số chẵn")
  else:
    # Nếu không phải số chẵn (là số lẻ), in ra thông báo tương ứng.
    print(x, "là số lẻ")
```

```python
#W1A5
# Khởi tạo một danh sách (list) các chuỗi
animals = ["cat", "dog", "cat", "bird"]
# Khởi tạo biến 'count' với giá trị ban đầu là 0
# Biến này sẽ đếm số lần lặp
count = 0
# Vòng lặp duyệt qua từng phần tử trong danh sách 'animals'
for a in animals:
  # Tăng biến 'count' lên 1 sau mỗi lần lặp
  count += 1
# Sau khi vòng lặp kết thúc, in ra giá trị cuối cùng của 'count'
print("Số phần tử trong danh sách là:", count)
```

```python
#W1A7
# Yêu cầu người dùng nhập một số.
# Giá trị nhập vào ban đầu là chuỗi (string).
# đổi thành nhập số để có thể thực hiện phép toán.
num = int(input())
# Dùng câu lệnh if-else để kiểm tra tính chẵn/lẻ.
# Toán tử % là chia lấy dư.
# Toán tử == là so sánh bằng.
# Điều kiện "num % 2 == 0" kiểm tra xem số có chia hết cho 2 không.
if num % 2 == 0:
    # Dòng này bị thụt lề, chỉ chạy nếu điều kiện trên đúng.
    print("Số chẵn")
# else phải kết thúc bằng dấu hai chấm.
else:
    # Dòng này bị thụt lề, chỉ chạy nếu điều kiện trên sai.
    print("Số lẻ")
```

```python
#W1A8
# Vòng lặp 'for' sẽ lặp lại một số lần cụ thể.
# 'range(3)' tạo ra một chuỗi số từ 0 đến 2 (0, 1, 2).
# Vòng lặp sẽ chạy ba lần, với 'i' lần lượt nhận các giá trị này.
for i in range(3):
  # Dòng này bị thụt lề, nên nó nằm trong vòng lặp.
  # Nó in ra thông báo thể hiện lần học hiện tại.
  # Vì 'i' bắt đầu từ 0, việc cộng thêm 1 giúp số lần đếm bắt đầu từ 1.
  # Kết quả sẽ là "AI đang học lần 1", rồi "AI đang học lần 2", và cuối cùng là "AI đang học lần 3".
  print("AI đang học lần", i + 1)
# Dòng này không bị thụt lề, nên nó nằm ngoài vòng lặp.
# Nó chỉ được thực thi một lần duy nhất, sau khi vòng lặp đã chạy xong cả ba lần.
print("Huấn luyện xong!")
```

```python
#W1A9
# Khởi tạo một vòng lặp 'for'
# Vòng lặp này sẽ duyệt qua từng phần tử trong danh sách ["cat", "dog", "fish"]
for x in ["cat", "dog", "fish"]:
  # Dòng code này nằm trong vòng lặp, nên nó sẽ được thực thi ba lần.
  # Nó in ra chuỗi "Dự đoán con vật:" và giá trị hiện tại của biến x.
  # Ví dụ: lần đầu tiên, x sẽ là "cat", dòng này sẽ in ra "Dự đoán con vật: cat"
  print("Dự đoán con vật:", x)
```
