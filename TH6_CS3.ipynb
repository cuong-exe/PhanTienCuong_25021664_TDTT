```python
#W6A1
numbers = list(map(int, input().split()))
no_dup = []
nums = {}
for x in numbers:
    if x not in nums:
        no_dup.append(x)
        nums[x] = 1
print(no_dup)
```

```python
#W6A2
nums = list(map(int, input().split()))
for i in range(1, len(nums)):
    nums[i] += nums[i - 1]
print(nums)
```

```python
#W6A3
t = tuple(map(int, input().split()))
k = int(input())
k = k % len(t)
new_t = t[k:] + t[:k]
print(new_t)
```

```python
#W6A4
pairs = input().split()
d = {}
for p in pairs:
    key, val = p.split(':')
    if key not in d:
        d[key] = []
    d[key].append(val)
print(d)
```

```python
#W6A5
nums = list(map(int, input().split()))
d = {'positives': 0, 'negatives': 0, 'zeros': 0}
for x in nums:
    if x > 0:
        d['positives'] += 1
    elif x < 0:
        d['negatives'] += 1
    else:
        d['zeros'] += 1
print(d)
```

```python
#W6A8
strs = list(input().split())
d = {}
for s in strs:
    if s not in d:
        d[s] = 0
    d[s] += 1
print(d)
```

```python
#W6A9
d1 = eval(input())
d2 = eval(input())
merged = d1.copy()
for key, val in d2.items():
    if key in merged:
        merged[key] += val
    else:
        merged[key] = val
sorted_d = dict(sorted(merged.item()))
print(sorted_d)
```

```python
#W6A10
nums = list(map(int, input().split()))
k = int(input())
dic = {}
ans = []
for x in nums:
    if k - x in dic:
        ans.append((k - x, x))
    dic[x] = 1
for p in ans:
    print(p)
```

```python
#W6A11
t = tuple(map(int, input().split()))
even = []
odd = []
for x in t:
    if x % 2:
        odd.append(x)
    else:
        even.append(x)
t1 = tuple(even)
t2 = tuple(odd)
print(t1)
print(t2)
```

```python
#W6A12
nums = list(map(int, input().split()))
dic = {}
for x in nums:
    if x not in dic:
        dic[x] = 0
    dic[x] += 1
val = 0
key = -1
for k, v in dic.items():
    if v > val:
        key = k
        val = v
    elif v == val:
        key = min(key, k)
print(key, val)
```
