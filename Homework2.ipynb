```python
#Bai3
a, b, c = map(float, input("Nhập 3 số a, b,c: ").split())
if (a > 0 and b > 0 and c > 0 and a + b > c and b + c > a and a + c > b):
    if (a == b == c):
        print("Tam giác đều")
    elif (a == b or a == c or b == c):
        print("Tam giác cân")
    elif ( a * a + b * b == c * c or a * a + c * c == b * b or b * b + c * c == a * a):
        print("Tam giác vuông")
    else:
        print("Là Tam giác")
    p = (a + b + c) / 2
    s = p * (p - a) * (p - b) * (p - c)
    s = s ** 0.5
    print(f'Diện tích tam giác là: {s}', f'Chu vi là {a + b + c}', sep = '\n')
else: 
    print("Không phải là 3 cạnh của một tam giác")
```

```python
#Bai4
a, b = map(float, input("Nhập hai số a, b cho phương trình ax + b = 0: ").split())
if (a == 0 and b != 0):
    print("Phương trình vô nghiệm")
elif (a == 0 and b == 0):
    print("Phương trình có vô số nghiệm")
else:
    print(f'Phương trình có nghiệm duy nhất: x = {-b / a}')
```

```python
#Bai5
a, b, c = map(float, input("Nhập ba số a, b, c cho phương trình ax^2 + bx + c = 0: ").split())
if (a == 0):
    if (b == 0 and c != 0):
        print("Phương trình vô nghiệm")
    elif (b == 0 and c == 0):
        print("Phương trình có vô số nghiệm")
    else:
        print(f'Phương trình có một nghiệm là: {-c / b}')
else:
    delta = b * b - 4 * a * c
    if (delta < 0):
        print("Phương trình vô nghiệm")
    elif (delta == 0):
        print(f'Phương trình có nghiệm kép: x = {-b/2 * a}')
    else:
        x1 = (-b + delta ** 0.5) / (2 * a)
        x2 = (-b - delta ** 0.5) / (2 * a)
        print(f'Phương trình có hai nghiệm phân biệt x1 = {x1} và x2 = {x2}')
```
