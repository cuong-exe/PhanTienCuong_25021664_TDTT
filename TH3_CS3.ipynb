```python
#W3A16
x = float(input())
if x >= 0:
    down = int(x)
else:
    if x != int(x):
        down = int(x) - 1
    else:
        down = int(x)
if x >= 0:
    up = int(x) + (1 if x != int(x) else 0)
else:
    up = int(x)
decimal = x - int(x)
if x >= 0:
    nearest = int(x) + (1 if decimal >= 0.5 else 0)
else:
    nearest = int(x) - (1 if decimal <= -0.5 else 0)
print(f'lam tron xuong {down}')
print(f'lam tron len {up}')
print(f'so nguyen gan nhat {nearest}')
```
